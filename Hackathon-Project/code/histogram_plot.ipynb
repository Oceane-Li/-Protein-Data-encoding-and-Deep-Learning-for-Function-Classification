# Histrogram plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the data
data = pd.read_csv('eval_result.txt', sep='  ')
data["Code"] = pd.read_csv('code.txt', sep='  ')
data
```

```python
#plot the f1 score
plt.figure(figsize=(20,20))
plt.bar(data['Code'], data['f1-score'], label='f1')
plt.xlabel('InterPro')
plt.xticks(rotation=90)
plt.ylabel('F1 score')
plt.title('F1 score over InterPro for deep learning model')
```
